양의 정수 x를 입력 받아 제곱근의 근사값의 결과를 반환하는 함수를 작성하세요.

sqrt() 사용 금지



```python
# 정답
def bisection(num):
    a = 1
    b = num
    c = 0
    
    for i in range(11):
        c = (a+b)/2
        if c**2<num:
            a = c
        else:
            b = c
    
    return f'{a} < 루트 {num} < {b}'
    
    

print(bisection(2))
```
