```python
# 
def bisection(num):
    a = 1
    b = num
    c = 0
    
    for i in range(11):
        c = (a+b)/2
        if c**2<num:
            a = c
        else:
            b = c
    
    return f'{a} <  {num} < {b}'
    
    

print(bisection(2))
```
